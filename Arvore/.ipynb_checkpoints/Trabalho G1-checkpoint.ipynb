```python
class Tree : 
    
    def __init__(self):
        self.item = None
        self.esquerda = None
        self.direita = None
    
    def insert(self,item):
        if self.item == None: # verifiaca se há elemento na raiz
            self.item = item
            return
        elif item < self.item:
            if self.esquerda == None:
                self.esquerda = Tree()
            return self.esquerda.insert(item)
        else:
            if self.direita == None:
                self.direita = Tree()
            return self.direita.insert(item)

    def regra2(self):
        if self.direita != None :
            return self.direita.regra2()
        elif self.esquerda != None :
            print("Primeiro",self.item)
            value = self.item
            self.item = self.esquerda.item
            self.direita = self.esquerda.direita
            self.esquerda = self.esquerda.esquerda
            print("Segundo",self.item)
            return value
        else :
            value = self.item
            self = None
            return value

    def delete(self, item):
        print("oi")
        if self.item == item:
            if self.esquerda == None and self.direira == None :
                self.item = None
            else:
                value = self.item
                print("Item :",self.item)
                print("Fun :", self.esquerda.regra2())
                self.item = self.esquerda.regra2()
                print("Value > ",value)
                return value
        elif item < self.item:
            self.item = self.esquerda.delete(item)
        else:
            self.item = self.direita.delete(item)
        
        return

    def search(self, item):
        if self.item == item: # verifiaca se há elemento na raiz
            return True
        elif item < self.item:
            if self.esquerda == None:
                return False
            return self.esquerda.search(item)
        else:
            if self.direita == None:
                return False
            return self.direita.search(item)
        
    def inOrdem(self):
        if self.item != None:
            if self.esquerda != None :
                self.esquerda.inOrdem()#E
            print(self.item)           #R
            if self.direita != None :
                self.direita.inOrdem() #D
    
    def posOrdem(self) :
        if self.item != None:
            if self.esquerda != None :
                self.esquerda.inOrdem()#E
            if self.direita != None :
                self.direita.inOrdem() #D
            print(self.item)           #R

    def preOrdem(self) :
        if self.item != None:
            print(self.item)           #R
            if self.esquerda != None :
                self.esquerda.inOrdem()#E
            if self.direita != None :
                self.direita.inOrdem() #D
    
        
arvore = Tree()
arvore.insert(10)
arvore.insert(5)
arvore.insert(4)
arvore.insert(7)
arvore.insert(6)
arvore.insert(2)
arvore.insert(13)
arvore.insert(12)
arvore.insert(14)
arvore.insert(15)

# Mostrar em preOrdem
print("Mostrar em preOrdem")
arvore.preOrdem()
# Mostrar em ordem
print("Mostrar em ordem")
arvore.inOrdem()
# Mostrar em posOrdem
print("Mostrar em posOrdem")
arvore.posOrdem()
print("Buscar numero 3")
print(arvore.search(3))
print("Buscar numero 18")
print(arvore.search(18))

print("Delete 7")
arvore.delete(7)
print("Mostrar em preOrdem")
arvore.preOrdem()
```
